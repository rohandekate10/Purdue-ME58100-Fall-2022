```python
import numpy as np
# Define the function
def f(t,y):
    return ((np.sin(y) - np.exp(t))/np.cos(t))
```

### h = 0.5

```python
# Forward Euler Method

def euler_forward(a,b,h,y0):
    Nf=(b-a)/h
    tf=np.linspace(a,b,int(round(Nf+1)))
    yf=np.zeros(int(round(Nf+1)))
    tf[0]=a
    yf[0]=y0
    for i in range(1, int(round(Nf+1))):
        yf[i]= yf[i-1]+h*f(tf[i-1],yf[i-1])
        
    return Nf,tf,yf

# Euler Method with h = 0.5
Nf_1,tf_1,yf_1=euler_forward(0, 1, 0.5, 0)
#print(Nf_1,tf_1,yf_1)
print(f"Number of Points: {Nf_1}")
print(f"Time Steps: {tf_1}")
print(f"Estimated Solutions for each Time Step: {yf_1}")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(tf_1,yf_1,'.',label='Forward Euler Method with $h = 0.5$')
plt.title('Estimated Solution $y(t_{i})$ vs. Time using Forward Euler Method with $h = 0.5$')
plt.xlabel('Time $(t)$')
plt.ylabel('Estimated Solution $y(t_{i})$')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
# Define the function
def f(t,y):
    return ((np.sin(y) - np.exp(t))/np.cos(t))

def df(t,y):
    return np.cos(y)/np.cos(t)
```

```python
def euler_backward(a,b,h,y0):
    Nb=(b-a)/h
    tb=np.linspace(a,b,int(round(Nb+1)))
    yb=np.zeros(int(round(Nb+1)))
    g = np.zeros(10)
    dg = np.zeros(10)
    z = np.zeros(10)
    tb[0]=a
    yb[0]=y0
    for i in range(int(round(Nb))):
        z[0] = yb[i]
        for j in range(0,9):
            g[j] = z[j] - yb[i] - h*f(tb[i+1],z[-1])
            dg[j] = 1 - h*df(tb[i+1],z[-1])
            z[j+1] = z[j] - g[j]/dg[j]
            
        yb[i+1] = z[-1]
    return Nb,tb, yb

Nf_5,tf_5,yb_1=euler_backward(0, 1, 0.02, 0)
print(f"Number of Points: {Nf_5}")
print(f"Time Steps: {tf_5}")
print(f"Estimated Solutions for each Time Step: {yb_1}")
```

```python
import numpy as np
# Define the function
def f(t,y):
    return ((np.sin(y) - np.exp(t))/np.cos(t))
```

```python
def RK2Method(a,y0,b,h):
    n = int((b - a) / h)+1
    print(f"Number of Time Steps: {n}")
    y = []
    y.append(y0)
    t = np.linspace(a,b,n)
    
    for idx in range(0,n-1):
        k1 = f(t[idx],y[idx])
        k2 = f(t[idx]+h, y[idx]+h*k1)
        
        y_new = y[idx]+(h/2)*(k1+k2)
        y.append(y_new)
    
    return y,t

y_sol,t = RK2Method(0,0,1,0.02)
print(y_sol)
```

```python
plt.plot(t,y_sol,'.', label = '2nd Order Runge Kutta Approximate Solution')
plt.title('Estimated Solution $y(t_{i})$ vs. Time for $h = 0.02$')
plt.xlabel('Time $(t)$')
plt.ylabel('Estimated Solution $y(t_{i})$')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def f1(t, y1, y2): 
    return 3*y1*(1-y2)

def f2(t, y1, y2): 
    return -5*y2*(1-y1)
```

```python
def calc_h(a,b,N):
    return (b-a)/(N-1)

def RK2Method(a,y10,y20,b,n):
    h = calc_h(a,b,n)
    print(f"Number of Time Steps N: {n}")
    print(f"Time Step Size h: {h}")
    y1 = []
    y2 = []
    y1.append(y10)
    y2.append(y20)
    
    t = np.linspace(a,b,n)
    
    for idx in range(0,n-1):
        k11 = f1(t[idx], y1[idx], y2[idx])
        k21 = f2(t[idx], y1[idx], y2[idx])
        
        k12 = f1(t[idx]+h, y1[idx]+h*k11, y2[idx]+h*k21)
        k22 = f2(t[idx]+h, y1[idx]+h*k11, y2[idx]+h*k21)
        
        y_new1 = y1[idx]+(h/2)*(k11+k12)
        y1.append(round(y_new1,8))
        
        y_new2 = y2[idx]+(h/2)*(k21+k22)
        y2.append(round(y_new2,8))

    #plt.plot(t,y1,'*', label='y1')
    #plt.plot(t,y2,'v', label='y2')
    #plt.title(f"Approximate Solution for N = {n}")
    #plt.xlabel('t')
    #plt.ylabel('y')
    #plt.legend()
    #plt.grid()
    #plt.show()
    return y1,y2,t

y1_n10, y2_n10, t_n10 = RK2Method(0,3,5,8,10)
print(y1_n10, y2_n10, t_n10)
```

```python
y1_n20, y2_n20, t_n20 = RK2Method(0,3,5,8,20)
print(y1_n20, y2_n20, t_n20)
```

* The common times are the 0 and the 8.
* At t = 0, the estimated solution for N and 2N is the same.
* So it only makes sense to compare solutions at t = 8 for a precision of 10e-6.
* Note - For a precision of 10e-8, the solution isn't converging as the method for calculating error is not specified. The absolute difference between the error diverges for smaller precision values and it takes longer for the function to output a result. Hence a precision of 10e-6 is choes and the results are calculated below.

```python
def get_common_solution2(sol1, sol2):
    tol = 10**(-8)
    diff = np.abs(sol1[-1] - sol2[-2])
    print(f"The error at the last time step is: {diff:.8f}")  
    return diff

diff1 = get_common_solution2(y1_n10,y1_n20)
diff2 = get_common_solution2(y2_n10,y2_n20)
```

```python
n1 = 10
a = 0
y10 = 3
y20 = 5
b = 8
tol = 10e-6

iteration = range(1,100)
for i in iteration:
    print(f"Iteration #: {i}")
    
    n2 = n1*2
    y1_1, y2_1, t_1 = RK2Method(a,y10,y20,b,n1)
    y1_2, y2_2, t_2 = RK2Method(a,y10,y20,b,n2)
    
    diff1 = get_common_solution2(y1_1,y1_2)
    diff2 = get_common_solution2(y2_1,y2_2)
    print()
    
    n1 = n1*2
    
    if (diff1<=tol) and (diff2<=tol):
        print(f"Solution Converged with errors {diff1, diff2}")
        break
    else:
        continue
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def f1(t, y1, y2): 
    return 3*y1*(1-y2)

def f2(t, y1, y2): 
    return -5*y2*(1-y1)

def calc_h(a,b,N):
    return (b-a)/(N-1)

def RK4Method(a,y10,y20,b,n):
    h = calc_h(a,b,n)
    print(f"Number of Time Steps: {n}")
    print(f"Time Step Size h = {h}")
    y1 = []
    y2 = []
    y1.append(y10)
    y2.append(y20)
    
    t = np.linspace(a,b,n)
    
    for idx in range(0,n-1):
        k11 = f1(t[idx], y1[idx], y2[idx])
        k21 = f2(t[idx], y1[idx], y2[idx])
        
        k12 = f1(t[idx]+(h/2), y1[idx]+k11*(h/2), y2[idx]+k21*(h/2))
        k22 = f2(t[idx]+(h/2), y1[idx]+k11*(h/2), y2[idx]+k21*(h/2))
        
        k13 = f1(t[idx]+(h/2), y1[idx]+k12*(h/2), y2[idx]+k22*(h/2))
        k23 = f2(t[idx]+(h/2), y1[idx]+k12*(h/2), y2[idx]+k22*(h/2))
        
        k14 = f1(t[idx]+h, y1[idx]+h*k13, y2[idx]+h*k23)
        k24 = f2(t[idx]+h, y1[idx]+h*k13, y2[idx]+h*k23)
        
        y_new1 = y1[idx]+(h/6)*(k11+2*k12+2*k13+k14)
        y1.append(y_new1)
        
        y_new2 = y2[idx]+(h/6)*(k21+2*k22+2*k23+k24)
        y2.append(y_new2)

    plt.plot(t,y1, label='y1')
    plt.plot(t,y2, label='y2')
    plt.title(f"4th Order Runge-Kutta Method Approximate Solutions for N = {n}")
    plt.xlabel('t')
    plt.ylabel('y')
    plt.legend()
    plt.grid()
    plt.show()
    return y1,y2,t

y1_n10, y2_n10, t_n10 = RK4Method(0,3,5,8,610)
```

## (a)

Let 

$u_1$ = $y$

$u_2$ = $y'$

$u_3$ = $y''$

Then,

$u_{1}'$ = $y'$ = $u_2$

$u_{2}'$ = $y''$ = $u_3$

$u_{3}'$ = $y'''$

Hence,

$u_{3}' = \frac{-1}{4}(u_{1}u_{3}+2u_{2}^2)$

For the initial conditions,

$u_{1}(0) = 0$

$u_{2}(0) = 0$

$u_{3}(0) = \frac{2^\frac{5}{2}}{9}$

## (b)

```python
import numpy as np
import matplotlib.pylab as plt


def f1(t,u1,u2,u3): return u2
def f2(t,u1,u2,u3): return u3
def f3(t,u1,u2,u3): return (-1/4)*(u1*u3+2*u2**2)
```

```python
def euler_forward(ti,tf,h,y10,y20,y30):
    print(f"Time Step size h: {h}")
    N=(tf-ti)/h
    print(f"N: {N}")
    t=np.linspace(ti,tf,int(round(N+1)))
    
    y1=np.zeros(int(round(N+1)))
    y2=np.zeros(int(round(N+1)))
    y3=np.zeros(int(round(N+1)))
    #y4=np.zeros(int(round(N+1)))
    
    t[0]=ti
    
    y1[0]=y10
    y2[0]=y20
    y3[0]=y30
    #y4[0]=y40
    
    for i in range(1, int(round(N+1))):
        y1[i]= y1[i-1]+h*f1(t[i-1],y1[i-1],y2[i-1],y3[i-1])
        y2[i]= y2[i-1]+h*f2(t[i-1],y1[i-1],y2[i-1],y3[i-1])
        y3[i]= y3[i-1]+h*f3(t[i-1],y1[i-1],y2[i-1],y3[i-1])
        #y4[i]= y4[i-1]+h*f4(t[i-1],y1[i-1],y2[i-1],y3[i-1],y4[i-1],w)
    return N,t,y1,y2,y3

ti = 0
tf = 30
h = 0.15
y10 = 0
y20 = 0
y30 = (2**(5/2))/9
#y40 = 0
#w = 1
N_eu,t_eu,u1_eu,u2_eu,u3_eu = euler_forward(ti,tf,h,y10,y20,y30)

print(f"u1: {u1_eu}")
print(f"u2: {u2_eu}")
print(f"u3: {u3_eu}")
```

## (c)

```python
def f1(t,u1,u2,u3): return u2
def f2(t,u1,u2,u3): return u3
def f3(t,u1,u2,u3): return (-1/4)*(u1*u3+2*u2**2)

def calc_N(a,b,h):
    return round(((b-a)/h)+1)

def RK4Method(a,y10,y20,y30,b,h):
    n = calc_N(a,b,h)
    print(f"Number of Time Steps: {n}")
    print(f"Time Step Size h = {h}")
    y1 = []
    y2 = []
    y3 = []
    y1.append(y10)
    y2.append(y20)
    y3.append(y30)
    
    t = np.linspace(a,b,n)
    
    for idx in range(0,n-1):
        k11 = f1(t[idx], y1[idx], y2[idx], y3[idx])
        k21 = f2(t[idx], y1[idx], y2[idx], y3[idx])
        k31 = f3(t[idx], y1[idx], y2[idx], y3[idx])
        
        k12 = f1(t[idx]+(h/2), y1[idx]+k11*(h/2), y2[idx]+k21*(h/2), y3[idx]+k31*(h/2))
        k22 = f2(t[idx]+(h/2), y1[idx]+k11*(h/2), y2[idx]+k21*(h/2), y3[idx]+k31*(h/2))
        k32 = f3(t[idx]+(h/2), y1[idx]+k11*(h/2), y2[idx]+k21*(h/2), y3[idx]+k31*(h/2))
        
        k13 = f1(t[idx]+(h/2), y1[idx]+k12*(h/2), y2[idx]+k22*(h/2), y3[idx]+k32*(h/2))
        k23 = f2(t[idx]+(h/2), y1[idx]+k12*(h/2), y2[idx]+k22*(h/2), y3[idx]+k32*(h/2))
        k33 = f3(t[idx]+(h/2), y1[idx]+k12*(h/2), y2[idx]+k22*(h/2), y3[idx]+k32*(h/2))
        
        k14 = f1(t[idx]+h, y1[idx]+h*k13, y2[idx]+h*k23, y3[idx]+h*k33)
        k24 = f2(t[idx]+h, y1[idx]+h*k13, y2[idx]+h*k23, y3[idx]+h*k33)
        k34 = f3(t[idx]+h, y1[idx]+h*k13, y2[idx]+h*k23, y3[idx]+h*k33)
        
        y_new1 = y1[idx]+(h/6)*(k11+2*k12+2*k13+k14)
        y1.append(y_new1)
        
        y_new2 = y2[idx]+(h/6)*(k21+2*k22+2*k23+k24)
        y2.append(y_new2)
        
        y_new3 = y3[idx]+(h/6)*(k31+2*k32+2*k33+k34)
        y3.append(y_new3)

    return y1,y2,y3,t

a = 0
b = 30
h = 0.15
y10 = 0
y20 = 0
y30 = (2**(5/2))/9

u1_rk4, u2_rk4, u3_rk4, t_rk4 = RK4Method(a,y10,y20,y30,b,h)

print(f"u1: {u1_rk4}")
print(f"u2: {u2_rk4}")
print(f"u3: {u3_rk4}")
```

```python
plt.plot(t_rk4,u1_rk4, label='u1')
plt.plot(t_rk4,u2_rk4, label='u2')
plt.plot(t_rk4,u3_rk4, label='u3')
plt.title(f"4th Order Runge-Kutta Method Approximate Solutions")
plt.xlabel('t')
plt.ylabel('$u_{i}(t_{i})$')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pylab as plt


def f1(t,u1,u2): return u2
def f2(t,u1,u2,k): return ((-np.exp(u1))/k)*(1+2*u2)
```

```python
def calc_N(a,b,h):
    return round(((b-a)/h))

def RK2Method(a,y10,y20,b,h,k):
    n = calc_N(a,b,h)
    print(f"Number of Time Steps: {n}")
    print(f"Time Step Size h = {h}")
    print(f"$\kappa = {k}$")
    y1 = []
    y2 = []
    y1.append(y10)
    y2.append(y20)
    
    t = np.linspace(a,b,n)
    
    for idx in range(0,n-1):
        k11 = f1(t[idx], y1[idx], y2[idx])
        k21 = f2(t[idx], y1[idx], y2[idx],k)
        
        k12 = f1(t[idx]+h, y1[idx]+h*k11, y2[idx]+h*k21)
        k22 = f2(t[idx]+h, y1[idx]+h*k11, y2[idx]+h*k21,k)
        
        y_new1 = y1[idx]+(h/2)*(k11+k12)
        y1.append(round(y_new1,8))
        
        y_new2 = y2[idx]+(h/2)*(k21+k22)
        y2.append(round(y_new2,8))
        
    #plt.plot(t,np.exp(y1), label='Population Growth')
    plt.plot(t,y1, label='Population Growth')
    #plt.plot(t,y2,'v', label='y2')
    plt.title(f"Population vs. Time for h={h}")
    plt.xlabel('t')
    plt.ylabel('Population')
    plt.legend()
    plt.grid()
    plt.show()
    return y1,y2,t

a=0
u0 = 0.5
k = 0.01
y10=np.log(u0)
y20=(1-u0)/k
b=10
h=0.1

y1, y2, t = RK2Method(a,y10,y20,b,h,k)
```
